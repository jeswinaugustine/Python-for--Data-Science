```python
import matplotlib.pyplot as plt 
import numpy as np
```

```python
%matplotlib inline
```

```python
# to show the plot 
#plt.show()
```

```python
# Example
```

```python
x = np.linspace(0,5,11)
y = x **2
```

```python
x
```

```python
y
```

```python
## Functional method of plot a matlibplot 
plt.plot(x,y)
plt.show()
```

```python
# Adding labels
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('Title')
plt.plot(x,y)
```

```python
## Multi-level plots 
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# Object oriented method
```

```python
fig = plt.figure() ## resembles a blank canvas

axes = fig.add_axes([1,1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X LABEL')
axes.set_ylabel('Y LABEL')
axes.set_title('TITLE')
```

```python
fig  = plt.figure ()
axes1 = fig.add_axes([0.0,0.0,0.8,0.8])
axes2 = fig.add_axes([0.1,0.5,0.3,0.3])
axes1.plot(x,y)
axes2.plot(y,x)
```

```python
# you can specifgy number of rows and colows 
fig,axes = plt.subplots(nrows=1,ncols=2)

#axes.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=5,ncols=4)
plt.tight_layout()
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)


for current_ax in axes:
    current_ax.plot(x,y)
```

```python
axes
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
#can be indexed as well 
axes[0].plot(x,y,"r")
axes[0].set_xlabel('X LABEL')
axes[0].set_ylabel('Y LABEL')
axes[0].set_title('TITLE 1')

axes[1].plot(y,x,"b")
axes[1].set_xlabel('X LABEL')
axes[1].set_ylabel('Y LABEL')
axes[1].set_title('TITLE 2')
```

## Figure size and DPI

```python
fig = plt.figure(figsize=(3,2))
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig = plt.figure(figsize=(8,4))
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,label='X')
ax.plot(y,x,label='1/X')
ax.set_xlabel('X LABEL')
ax.set_ylabel('Y LABEL')
ax.set_title('TITLE 2')
#adding legend
ax.legend(loc=0)
```

```python
fig.savefig("my_pic",dpi=200)
```

```python
#color 
fig = plt.figure(figsize=(8,4))
ax = fig.add_axes([0,0,1,1])
#ax.plot(x,y,label='X',color='#900000')
#ax.plot(x,y,label='X',color='purple',linewidth=3,linestyle='-')
#Marker
ax.plot(x,y,label='X',color='purple',linewidth=3,linestyle='-',marker='o',markersize=10,markerfacecolor='yellow')
ax.set_xlabel('X LABEL')
ax.set_ylabel('Y LABEL')
ax.set_title('TITLE 2')
#adding legend
ax.legend(loc=0)
```

```python
#Marker
fig = plt.figure(figsize=(8,4))
ax = fig.add_axes([0,0,1,1])
#ax.plot(x,y,label='X',color='#900000')
ax.plot(x,y,label='X',color='purple',linewidth=3,linestyle='-')
ax.set_xlabel('X LABEL')
ax.set_ylabel('Y LABEL')
ax.set_title('TITLE 2')
#adding legend
ax.legend(loc=0)
```

Special Plot Types

There are many specialized plots we can create, such as barplots, histograms, scatter plots, and much more. Most of these type of plots we will actually create using seaborn, a statistical plotting library for Python. But here are a few examples of these type of plots:



```python

plt.scatter(x,y)

from random import sample

data = sample(range(1, 1000), 100)

plt.hist(data)
```
